# Problem 1

```python
flight_airports = flights.select('ORIGIN_AIRPORT','DESTINATION_AIRPORT')
flight_airports
```

This table is used to select only the airports of each flight for the later purpose of counting the number of flights in and out of each airport.

```python
origin_airport_count = flight_airports.group('ORIGIN_AIRPORT').sort('count',descending=True)
origin_airport_info = origin_airport_count.join('ORIGIN_AIRPORT', new_airport_info,'IATA_CODE')
origin_airport_info
```

The above code snippet counts the number of times an **origin airport** appears in the flight_airports table created earlier. Then, it is joined with the location information of each airport to later map all of the information

```python
destination_airport_count = flight_airports.group('DESTINATION_AIRPORT').sort('count',descending=True)
destination_airport_count
```

The above code makes a table which has the number of times an airport occurs in the **destination airport** column of the flights.csv file.

```python
destination_airport_info = destination_airport_count.join('DESTINATION_AIRPORT',new_airport_info,'IATA_CODE')
destination_airport_info
```

The above code joins the **destination airport** information with the location information of each airport for the purpose of mapping the data.

```python
# color information for origin table
count_colors = pd.cut(origin_airport_info.column('count'),
     bins = [0, 10000, 50000, 100000, 250000, 350000],
     labels = ['blue', 'green', 'yellow', 'orange',
     'red'])

origin_airport_counts = origin_airport_info.with_column('colors', count_colors)

# color information for destination table
count_colors = pd.cut(destination_airport_info.column('count'),
     bins = [0, 10000, 25000, 50000, 100000, 300000],
     labels = ['blue', 'green', 'yellow', 'orange',
     'red'])
# Add this newly created array of colors to the Table:
destination_airport_counts = destination_airport_info.with_column('colors', count_colors)
```

The above code sets up for making a color coded map to show the number of flights going out of an airport (origin_airport) and a map showing the number of flights going into an aiport (destination_airport)

```python
origin_map_info = origin_airport_counts.select('lat', 'long', 'ORIGIN_AIRPORT', 'count', 'colors')
origin_map_info.relabel('ORIGIN_AIRPORT', 'labels')

Map1_info = origin_map_info.with_column('areas', origin_airport_counts.column('count')*0.03)
Map1_info.show(5)
```

The above code sets up creating the map to show the origin airports.

```python
destination_map_info = destination_airport_counts.select('lat', 'long', 'DESTINATION_AIRPORT', 'count', 'colors')
destination_map_info.relabel('DESTINATION_AIRPORT', 'labels')

Map2_info = destination_map_info.with_column('areas', origin_airport_counts.column('count')*0.03)
Map2_info.show(5)
```

The above code sets up creating the map to show the destination airports.

```python
Map1_info = Map1_info.with_column('areas', Map1_info.column('count')*0.003)

Circle.map_table(Map1_info.select('lat', 'long', 'labels', 'colors', 'areas'))
```

The above code creates the visualization for origin airports.

```python
Map2_info = Map2_info.with_column('areas', Map2_info.column('count')*0.003)

Circle.map_table(Map2_info.select('lat', 'long', 'labels', 'colors', 'areas'))
```

The above code creates the visualization for destination airports.
