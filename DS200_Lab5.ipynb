# Problem 2

```python
# rename table columns
locations = locations.relabel('Lat', 'lat')
locations = locations.relabel('Long', 'long')
locations = locations.relabel('Campus', 'labels')

# reorder columns
locations = locations.select('lat','long','labels')

# map
Marker.map_table(locations)
```

# Problem 3

```python
all_info = demographics.join('Location', locations, 'labels')
all_info.labels
```

# Problem 4

```python
GroupedByCampus = all_info.group('Location')
GroupedByCampus.show()
```

# Problem 5


```python
location_population = GroupedByCampus.join('Location', locations, 'labels')
starts_map_data = location_population.select('lat', 'long', 'Location').with_columns(
    'colors', 'blue',
    'areas', location_population.column('count') * 0.3
)

Circle.map_table(starts_map_data.relabel('Location', 'labels'))

```
